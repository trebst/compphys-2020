Die Benutzung von Zufall im Computer spielt nicht nur in naturwissenschaftlichen Simulationsrechnungen eine große Rolle, sondern wird zunehmend auch in anderen Bereichen eingesetzt. In dieser speziellen Aufgabe, wollen wir die **Generierung von Landschaften** im Computer untersuchen. Grundsätzlich unterscheidet man hier zwischen zwei Arten digitaler Landschaften, die erste ist eine individuelle, artistische Gestaltung per Hand die besonders in **CGI** (Computer-Generated Images) verwendet wird, sowie zum anderen die **automatische und zufällige Levelgenerierung** in Computerspielen, die Sie z.b. von Minecraft kennen werden.



Grundproblem bei der automatischen Generierung von Landschaften, ist die Vereinbarkeit von *zufälligem* Rauschen und gleichzeitig einer *kontinuierlichen* Oberfläche, die zwar zufällige Strukturen, aber keine Diskontinuitäten aufweist. Über die letzten 40 Jahre wurde eine Vielzahl von Algorithmen zur zufälligen Generierung von Landschaften entworfen, von denen wir Ihnen in dieser Aufgabe zwei Beispiele vorstellen wollen. Beide Algorithmen haben gemeinsam, dass sie eine zweidimensionale Oberfläche generieren, die in Folgeschritten mit weiteren Verfahren (z.B. Erosion, Flüssigkeiten, Vegetation, etc.) verfeinert werden kann. 

### Diamond-Square Algorithmus

Der [**Diamond-Square Algorithmus**](https://de.wikipedia.org/wiki/Diamond-square_Algorithmus) wurde in den 1980er Jahren für die Generierung zweidimensionaler Höhenfelder entwickelt. Er basiert auf dem iterativen Verfeinern einer zuerst groben Höhenverteilung, bei dem in zwei abwechselnden Einzelschritten die Landschaft detaillierter wird.

Die Funktionsweise des Algorithmus lässt sich in folgender Grafik erklären:



Der Ausgangspunkt jeder Iteration ist eine Höhenverteilung über einem quadratischen Gitter. Für das Starten Ihres Algorithmus können Sie z.b. eine kleine 2x2 Matrix mit zufälligen Höhen initialisieren. Die eigentliche Diamond-Square Iteration besteht aus zwei Schritten:

1. Im *Diamond step* wird der diagonale Mittelpunkt jeder quadratischen Plaquette mit einem Wert gefüllt, welcher sich aus dem Mittelwert der umliegenden 4 Felder, sowie einer kleinen zufälligen Abweichung berechnet.
2. Im *Square step* werden die nun entstandenen Lücken gefüllt indem der Mittelwert entlang der x-/y-Richtung mit einer weiteren kleinen zufälligen Abweichung benutzt wird.

Nachdem beide Schritte vollzogen wurden, ist die resultierende Konfiguration erneut auf einem quadratischen Gitter definiert und kann als Ausgangspunkt für die nächste Iteration benutzt werden. Während den einzelnen Schritten dient der Mittelwert dazu, dass die generierte Verteilung stetig und glatt wird, während die zufällige Abweichung vom Mittelwert ein zufälliges Rauschen in die Landschaft induziert. Dabei sollte die Amplitude mit jeder Iteration reduziert werden.

### Perlin Noise

Ein fortgeschrittenerer Ansatz als der Diamond-Square Algorithmus basiert auf sogenanntem [Perlin Noise](https://de.wikipedia.org/wiki/Perlin-Noise), einer Form von korreliertem Rauschen, das ursprünglich von Ken Perlin für die Landschaftsgenerierung im Film Tron (1982) entwickelt wurde. Dabei wird eine zufällige Amplitude $h(x)$ für Stützstellen $x$ so gezogen, dass nahe Stützstellen $x$ und $x^*$ ähnliche Amplituden besitzen. Um dies zu gewährleisten, wird beim Perlin Noise nicht die Amplitude selbst zufällig bestimmt, sondern der lokale Gradient auf einem groben Grid von Punkten, zwischen denen letzten Endes die Amplituden interpoliert werden. Um mittels Perlin Noise eine Landschaft zu generieren, werden in der Regel Rausch-Bilder verschieden grober Grids überlagert.





Durch seine stark kontrollierbare Form und hohen Grad an individuellen Anpassungs-Möglichkeiten ist Perlin Noise heute ein Standardwerkzeug bei der zufälligen Generierung von Landschaften, und wird z.B. in der [Levelgenerierung von Minecraft](https://minecraft-de.gamepedia.com/Weltgenerierung) oder für die prozedurale Generierung von Textures verwendet. Im Folgenden wollen wir Ihnen den Algorithmus näher bringen und Ihnen ermöglichen, selber mit dem Algorithmus Landschaften zu erzeugen. Für ein weiterführendes Tutorial empfehlen wir z.b. [dieses Tutorial](https://flafla2.github.io/2014/08/09/perlinnoise.html) oder [dieses Tutorial](https://connect.unity.com/p/an-exploration-of-perlin-noise).

Die Erzeugung von Perlin Noise erfolgt in mehreren Schritten. Zuerst wird ein zufälliges Gradientenfeld erzeugt. Dies funktioniert wie folgt.

1. Es wird ein Grid von Gitterpunkten über den Raumbereich gelegt, in dem die zufälligen Amplituden gezogen werden sollen.
2. An jedem der Gitterpunkte wird ein zufälliger Gradientenvektor bestimmt. 

Für jeden Punkt innerhalb des Gitterbereichs, für den man eine Amplitude bestimmen möchte, wird folgende Prozedur benutzt.

1. Bestimme in welchem Gitterquadrat sich der Punkt befindet (wichtig ist hier die Kentniss über die 4 angrenzenden Gradienten an den Ecken)
2. Bestimme die Abstandsvektoren zu den Ecken an denen die Gradienten definiert sind
3. Bilde das Skalarprodukt der Abstandsvektoren mit den Gradienten, dies ergibt 4 Werte
4. Die berechneten 4 Werte werden jetzt gewichtet summiert. Dabei werden zuerst die oberen zwei Werte und die unteren zwei Werte linear mittels $A(\omega) = a (1-\omega) + b \omega$ interpoliert, und schließlich die resultierenden Mittelwerte ebenfalls linear interpoliert. Für die Kontinuität beim Interpolieren verwendet man die Funktion $\omega(t) = 6 t^5 - 15 t^4 + 10 t^3$ wobei $t$ die relative x- bzw. y-Position im Gitterfeld bezeichnet.
