<td style="border-style: none; width: 1%; text-align: right; font-size: 15px">[10 Punkte]</td></tr></table>


Den Plattenkondensator und sein **elektrisches Feld** haben Sie schon in der einführenden Vorlesung zur Experimentalphysik kennengelernt. Meist wird dabei die Annahme eines homogenen Feldes innerhalb des Kondensators gemacht, so dass die Feldlinien ein gerades Bündel innerhalb des Kondensators formen entsprechend der folgenden Abbildung. 


In dieser Aufgabe wollen wir überprüfen, in welchem Maße diese Annahme zutreffend ist. Dazu wollen wir das Feld des Plattenkondensators aus der Lösung der **Laplace-Gleichung**

\begin{equation}
\Delta\phi = 0
\end{equation}

numerisch exakt berechnen, etwa indem wir auf die in der Vorlesung vorgestellte **Relaxationsmethode** zurückgreifen. Die Randbedingungen seien dabei so formuliert, dass die beiden Kondensatorplatten auf einem Potential von $+1$ bzw. $-1$ und der Rand der Fläche auf einem Potential von 0 liegt. Um die letzte Randbedingung zu rechtfertigen, müssen die Platten ausreichend weit vom Rand entfernt sein. In der Praxis können Sie die in der folgenden Abbildung dargestellte Konfiguration verwenden:


### a) Implementation der Relaxation in Julia

Starten Sie indem Sie den in der Vorlesung vorgestellten Relaxations-Algorithmus für die Laplace-Gleichung mit den Randbedingungen aus der obigen Abbildung implementieren. Berechnen Sie das resultierende Potential (z.b. nach 10.000 Relaxationsschritten) und plotten Sie es mittels `imshow`.

### b) Gradienten und elektrisches Feld

Berechnen Sie dann das resultierende elektrische Feld als Gradient des berechneten Potentials. Um das Feld darzustellen, sollen Sie die PyPlot Funktion [`quiver`](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.quiver.html) verwenden. Untersuchen Sie die [Dokumentation](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.quiver.html) der Funktion und machen Sie sich mit Ihrer Benutzung vertraut. Plotten Sie dann das elektrische Feld des Plattenkondensators zusätzlich zum berechneten Potential. 

Inwiefern stimmt diese Feldkonfiguration mit der schematischen Darstellung der ersten Grafik überein?

### c) (Optional) Initialen als Platten

Als optionale Zusatzaufgabe können Sie probieren statt der normalen Plattenkondensatorplatten Ihre Initialen als Platten zu verwenden. Sie können dafür z.b. in einem Bildprogramm Ihrer Wahl Ihre Initialen als .png Bild erzeugen und mittels des PyPlot Befehls `imread` in Julia einlesen. Sie müssen dann nur noch basierend auf Ihrem Bild die Randbedingungen in jedem Relaxationsschritt anpassen. Wie sieht das elektrische Feld zwischen Ihren Namensinitialen aus?
