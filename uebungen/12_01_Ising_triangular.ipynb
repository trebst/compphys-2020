In dieser Aufgabe wollen wir das Ising Modell auf dem Dreiecksgitter betrachten. Sie kennen bereits aus der Vorlesung, sowie aus den Programmier-Tutorials das Ising Modell auf dem Quadratgitter. Ziel dieser Aufgabe ist es, Ihre Erkenntnisse aus dem Beispiel des Quadratgitters auf die Diskussion des Dreiecksgitters zu übertragen. 

Das Ising Modell ist ein klassisches Spin-Modell welches sogenannte Ising Spins auf einem unterliegenden Gitter beschreibt. Ising Spins können dabei die Werte $\sigma_\alpha = \pm 1$ annehmen. Der komplette Hamiltonian des Modells ist

\begin{equation}
    \mathcal{H} = -J \sum_{\langle \alpha,\beta \rangle} \sigma_\alpha \sigma_\beta
\end{equation}

wobei die Summe über nächste Nachbarn auf dem Dreiecksgitter ausgeführt wird (siehe unten stehende Abbildung). $\alpha$ und $\beta$  bezeichnen dabei die Indices der verschiedenen Spins und können z.B. durch die Indices in den Gitterkoordinaten ausgedrückt werden: $\alpha = (i,j)$.

Um das Dreiecksgitter zu implementieren, lässt sich (wie in unten stehender Skizze gezeigt) das Quadratgitter um weitere diagonale Verbindungen erweitern. Dies bedeutet, dass Sie Ihren Code aus den Programmier-Tutorials um weitere Verbindungen erweitern aber ansonsten verwenden können.



### a) Ising Modell auf dem Quadratgitter

Implementieren Sie zuerst das Ising Modell auf dem **Quadratgitter** genau wie in den Programmiertechniken. Plotten Sie ganz analog 

1. Energie gegen Temperatur
2. spezifische Wärme gegen Temperatur

jeweils für verschiedene Systemgrößen und zeigen Sie, dass der Ferromagnet ($J=+1$) und der Antiferromagnet ($J=-1$) genau den gleichen Kurvenverlauf und das gleiche Finite Size Scaling Verhalten besitzen.

*Hinweis: Beschränkten Sie sich auf gerade Systemgrößen $L$.*

### b) Ising Modell auf dem Dreiecksgitter

Die Korrespondenz des Ferro- und Antiferromagneten auf dem Quadratgitter ist nicht zufällig, sondern Konsequenz einer Gittersymmetrie. Das Quadratgitter ist bipartit, d.h. man kann es in genau zwei Mengen Gitterplätze (A und B) teilen, die durch Wechselwirkungen nur mit Plätzen der jeweilig anderen Gruppe verknüpft sind. Dadurch hat eine Transformation aller Spins auf Plätzen nur einer Gruppe (z.B. Gruppe A), welche die Spins umkehrt (also $\sigma_\alpha \to -\sigma_\alpha$), exakt die gleiche Auswirkung auf den Hamiltonian wie eine Transformation von $J \to -J$. Dies bedeutet insbesondere, dass sich für jede ferromagnetische Konfiguration mit Energie $E$ eine antiferromagnetische Konfiguration mit gleicher Energie nach dieser Vorschrift konstruieren lässt und somit der Ferro- und der Antiferromagnet das gleiche Verhalten in der Temperatur aufweisen.

Nun wollen wir uns das Dreiecksgitter anschauen, welches ein Beispiel für ein nicht-bipartites Gitter ist. Im Folgenden wollen wir zeigen, dass dies insbesondere bedeutet, dass der Ferro- und der Antiferromagnet stark unterschiedliche Eigenschaften besitzen.

Implementieren Sie auch hierfür eine Monte-Carlo Simulation nach dem Schema, dass Sie in den Programmiertechniken kennengelernt haben:

1. Zuerst sollen Sie Funktionen implementieren, die Ising Spins erzeugen und plotten.
2. Auf den bestehenden Konfigurationen wird eine Energie nach oben stehendem Hamiltonian definiert. Hierfür muss die Summe über Nachbarn eines Spins (i,j) implementiert werden. Verwenden Sie periodische Randbedingungen und die oben definierten Nachbarn im Dreiecksgitter.
3. Implementieren Sie das Metropolis-Update. Damit ist es nun möglich Markov-Ketten zu generieren.

Nachdem Sie Ihre Implementation getestet haben sollen Sie zeigen, dass der ferromagnetische ($J=+1$) und der antiferromagnetische ($J=-1$) Fall des Ising Modells auf dem Dreiecksgitter sich insbesondere in ihren Kurvenverläufen der Energie und spezifischen Wärme als Funktion der Temperatur unterscheiden.

Zeigen Sie die Unterschiede im Bezug auf den Phasenübergang, insbesondere:
1. die unterschiedliche kritische Temperatur
2. das unterschiedliche Verhalten für verschiedene Systemgrößen
