<td style="border-style: none; width: 1%; text-align: left; font-size: 25px; font-weight: bold; text-decoration: underline">Algorithmen im Fluss</td>
<td style="border-style: none; width: 1%; text-align: right; font-size: 15px">[12 Punkte]</td></tr></table>

In dieser Aufgabe möchten wir Ihnen einen Einblick in die Methode der **Flussdiagramme** geben. Dabei handelt es sich um eine Darstellung von Algorithmen, in der Sie Pfeilen von Anweisung zu Anweisung folgen und dabei den Ablauf des Algorithmus abbilden. Grundsätzliche lässt sich jedes Programm in einem Flussdiagramm darstellen und es ist besonders hilfreich einen unbekannten Algorithmus zuerst in einem Flussdiagramm zu untersuchen, bevor er implementiert wird.

Abgesehen von den Pfeilen, denen Sie durch das Flussdiagramm folgend, gibt es eine Reihe verschiedener Elemente, die Sie in Flussdiagrammen finden werden und verwenden können:



Grundsätzlich finden Sie die folgenden Elemente:
1. **Anweisungen** sind die Kernelemente Ihres Algorithmus. Hier wird festgelegt, *was* genau *wie* gemacht wird. Beispiele sind Variablen-Deklarationen oder Zuweisungen wie _"Setze x auf 3"_ oder aber auch generelle Anweisungen wie _"Ziehe meinen Mantel an"_.
2. Bei **Abfragen** teilen sich Stränge von Pfeilen auf. Hier *unterscheiden* Sie verschiedene Fälle. Beispiele für Abfragen sind _"Ist x größer als 6?"_ oder _"Ist es draußen kalt?"_. Grundsätzlich sind nicht nur die Antworten _"Ja"_ und _"Nein"_ möglich, sondern auch Zahlen oder ähnliches, deshalb wäre als Beispiel ebenfalls denkbar _"Welchen Wert hat x?"_.
3. **Ausgaben** ermöglichen Ihrem Algorithmus mit der Außenwelt (insbesondere Ihnen) zu kommunizieren. Grundsätzlich ist diese Box äquivalent zu einem `println` in Ihrem Julia-Code.
4. Werden einzelne Anweisungen sehr komplex oder verwenden Sie viel nicht-sichtbaren Code, werden die Boxen stattdessen als **Funktionsaufrufe** oder **Unterprogramme** bezeichnet und Sie zeichnen zusätzliche, vertikale Linien in die Box ein. Grundsätzlich ist der Unterschied zwischen Anweisungen und Funktionsaufrufen im Ermessen des Anwenders, aber Sie haben hier z.B. die Möglichkeit größere Anweisungen entsprechend zu kennzeichnen. Beispiele wären (im Gegensatz zu 1.) unter anderen _"Berechne die Eigenwerte von x"_ oder _"Suche meinen Mantel"_.
5. Pfeile dürfen zwar ineinander münden, aber nicht aus dem Nichts kommen oder im Nichts enden. Dafür gibt es die Boxen **Programmstart** und **Programmstop**, die sowohl den Anfang Ihres Algorithmus, als auch dessen Ende symbolisieren. Bei Funktionsdefinitionen wäre der Funktionsaufruf die Start-Box und das `return`-Statement die Stop-Box.

Im Folgenden sollen Sie drei verschiedene Flussdiagramme selber zeichnen. Hierfür können Sie entweder ein Grafikprogramm Ihrer Wahl verwenden und die Grafik als `.png` oder `.pdf` versenden, oder Sie zeichnen Ihr Diagramm auf ein Blatt Papier, fotografieren dieses und versenden schließlich das Bild.

### a) Ein Flussdiagramm zu "Hello World"

Zeichnen Sie zuerst ein Flussdiagramm, welches einen einfachen Algorithmus darstellt, der nach den Start den Satz _"Hello World!"_ ausgibt und dann endet.

### b) Schleifen in Flussdiagrammen

Wie Sie sicher bemerkt haben, gibt es in Flussdiagrammen zwar Boxen für *Anweisungen* und *Abfragen*, aber nicht für *Schleifen*. Der Grund dafür ist, dass sich Schleifen als Kombination von Anweisungen und Abfragen darstellen lassen. Hierfür erinnern wir uns, dass der Kern einer Schleife die wiederholte Ausführung des gleichen Code-Blocks ist.

in einem Flussdiagramm erreichen Sie eine erneute Ausführung des gleichen Code-Blocks dadurch, dass Sie mit Pfeilen einen geschlossenen Kreislauf bilden. Um die Anzahl der Wiederholungen zu kontrollieren, können Sie eine Variable `i` definieren, die als Zähler dient. Sie kennen diese Variable z.B. aus Ausdrücken wie `for i in 1:10`. Ein Durchlauf der Schleife besteht jetzt aus dem eigentlichen *Codeblock*, dem *Erhöhen* des Zählers `i` um 1 und dem *Testen* ob `i` bereits größer als eine obere Grenze ist. Sofern der Zähler seine obere Grenze noch nicht erreicht hat, wird der Fluss im Diagramm wieder vor den Code-Block geleitet, ansonsten verlässt der Fluss den Schleifenbereich. 

Benutzen Sie dieses Vorgehen um ein Flussdiagramm zu zeichnen, welches einen Algorithmus beschreibt, der die Zahlen von 1 bis 100 ausgibt.

### c) Der Weg in die Physik

Flussdiagramme werden nicht nur im Kontext von Programm-Code benutzt, sondern haben auch darüber hinaus eine gewisse Verbreitung, denn insgesamt lassen sich so alle Formen von komplexen Abläufen in einigermaßen übersichtlicher Form darstellen.

Zur Illustration dieses Punktes sollen Sie ein Flussdiagramm Ihres üblichen morgendlichen Weges zu den Kölner Physikalischen Instituten zeichnen (oder dem eines imaginären Studenten). Verwenden Sie in Ihrem Diagramm mindestens eine Abfrage, eine Schleife und eine Zuweisung.
