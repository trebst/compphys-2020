Im Folgenden möchten wir ein System betrachten, das im Englischen unter dem Namen [Chaos Game](https://en.wikipedia.org/wiki/Chaos_game) bekannt ist. Wir starten mit einer Anzahl von $N$ Eckpunkten und einem Anfangspunkt $(x,y)$ in der Ebene. Der Startpunkt wird später als eine der $N$ Ecken gewählt. Zusätzlich benötigen wir noch eine rationale Zahl $\alpha = p/q$ zwischen 0 und 1. Ausgehend von diesen Parametern soll eine Sequenz von Punkten mittels nachfolgender Update-Vorschrift erstellt werden.



In jedem Update-Schritt wird eine der Ecken zufällig ausgewählt. Dann wird der aktuelle Punkt $(x,y)$ um genau den Anteil $\alpha$ in Richtung dieser Ecke verschoben. Die so erreichte Position wird gespeichert und dient als Ausgangspunkt für den nachfolgenden Update-Schritt. So kann schlussendlich eine Sequenz von Positionen $(x_1,y_1), \dots, (x_n,y_n)$ generiert werden.

Implementieren Sie im Folgenden die beschriebene Update-Vorschrift und visualisieren Sie die berechneten Sequenzen von Punkten mittels `scatter` (und geeigneten Parametern) für folgende Szenarien:
1. Drei Ecken auf dem Einheitskreis, die ein gleichseitiges Dreieck bilden und Parameter $\alpha=1/2$
2. Fünf Ecken auf dem Einheitskreis, die ein gleichseitiges Pentagon bilden und Parameter $\alpha=1/2$
3. Gleiche Parameter wie bei 2., aber es darf nicht zweimal die gleiche Ecke nacheinander gewählt werden
4. Elf Ecken auf dem Einheitskreis, die ein gleichseitiges 11-Eck bilden, Parameter $\alpha=2/3$ und es darf erneut nicht die gleiche Ecke zweimal hintereinander gewählt werden

Sie dürfen auch gerne eigene Parameter-Sätze herumprobieren oder eine andere Art der Visualisierung (z.B. mittels `imshow`) ausprobieren.
