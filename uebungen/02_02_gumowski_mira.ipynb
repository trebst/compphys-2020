Der sogenannten **Gumowski-Mira Attraktors** ist ein **dynamisches System**, das einen zeitabhängigen Prozess beschreibt, dessen Entwicklung lediglich von der Wahl einiger Parameter sowie des Anfangszustands abhängt. Der **Phasenraum** des Systems ist zweidimensional, d.h. ein Zustand des Systems kann mit Koordinaten $(x, y)$ beschrieben werden. Außerdem besitzt das System drei Parameter $a, b, c$. Die Zeitentwicklung geschieht gemäß des folgenden Gesetz:

$\quad x_{n + 1} = y_n + a\cdot (1 + b\cdot y_n^2)\cdot y_n + f(x_n)$

$\quad y_{n + 1} = - x_n + f(x_{n + 1})$

wobei Funktion $f(x)$ gegeben ist durch

$\quad f(x) = c\cdot x + \dfrac{2\cdot(1 - c)\cdot x^2}{1 + x^2}$

Ziel der Aufgabe ist es eine Zeitreihe von `N` Punkten nach der obigen Iterationsvorschrift zu generieren und danach zu visualisieren.

### a) Plotten der Zeitreihen mittels `plot` oder `scatter`

Benutzen Sie die Parameter und Startkoordinaten

    a =  0.026             x1 = 0.1
    b = -0.034             y1 = 0.1
    c = -0.802095
und generieren Sie Listen aus x- und y-Werten mit jeweils 10.000 Elementen. Danach sollen Sie Ihre Daten plotten. 

*Hinweis: Versuchen Sie einen möglichst aussagekräftigen Plot zu erhalten, indem Sie mit den zusätzlichen Optionen für `plot(...)` bzw. `scatter(...)` herumspielen. Diese können Sie in der Dokumentation von `matplotlib` [hier](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html) und [hier](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html) finden. Dort finden Sie ebenfalls ein paar Beispiele für verschiedene Bilder, die mittels `plot` erzeugt wurden.*

### b) (Optional) Plotten der Zeitreihen mittels `imshow` als Bilder

Wie Sie sicher bei der Bearbeitung von Aufgabenteil a) bemerkt haben, ist eines der größten Probleme in der Darstellung der Zeitreihen, dass viele Datenpunkte sich gegenseitig überlappen. Dadurch verliert man den Überblick, in welcher Region des Plots sich wie viele Punkte befinden.

Abgesehen von den Darstellungsmöglichkeiten, die Sie in Teil a) implementiert haben, gibt es noch eine weitere Möglichkeit der Darstellung. Dabei wird über die Punkte ein Gitter aus Boxen gelegt und es wird gezählt, wie viele Punkte in welche Bereich des Gitters liegen. Diese Zählinformation kann dann mittels eines anderen `PyPlot` Befehls geplottet werden.



Gehen Sie also wie folgt vor:
1. Erzeugen Sie ein zweidimensionales Array, das mit `0` gefüllt ist mittels `zaehler = zeros(Int64, N,N)` wobei `N` die Anzahl der Boxen in die beiden Raumrichtungen beschreibt. Dieses Array soll als Zähler dienen und die Anzahl der Punkte in den jeweiligen Boxen speichern.
2. Prüfen Sie für jeden Punkt ihrer Zeitreihe in welcher Box er liegt und erhöhen Sie den Zähler dieser Box entsprechend
3. Benutzen Sie den Befehl `imshow(zaehler)`, um ihre Zaehlerstaende als Bild zu plotten. Auch hier können Sie in der [Dokumentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html) von `matplotlib` nach geeigneten Zusatzargumenten suchen, um Ihren Plot aussagekräftiger zu machen.
