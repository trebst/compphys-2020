In der nachfolgenden Aufgabe möchten wir Sie mit dem [Sierpinski-Dreieck](https://de.wikipedia.org/wiki/Sierpinski-Dreieck) vertraut machen, einem geometrischen zweidimensionalen Objekt, das fraktale und selbstähnliche Eigenschaften besitzt. Die besondere Eigenschaft, die wir in dieser Aufgabe erkunden möchten, ist die, dass es sich als Ergebnis eines eindimensionalen zellulären Automaten erzeugen lässt.



## Teil A: Das Sierpinksi-Dreieck

Das Sierpinski-Dreieck ist eine selbstähnliche geometrische Struktur in zwei Dimensionen. Sie besteht aus einem Dreieck, das immer wieder in sich selbst enthalten ist. Benutzen Sie die Selbstähnlichkeit des Sierpinski-Dreiecks und schreiben Sie eine Funktion, welche ein solches Gebilde zeichnet. Dabei sollen Sie wie folgt vorgehen:
1. Sie definieren eine Funktion `plot_dreieck`, welche ein *normales* Dreieck durch die übergebenen Punkte `A`, `B` und `C` zeichnet
2. Sie definieren eine Funktion `plot_sierpinski_dreieck`, die ein *Sierpinski-Dreieck* durch übergebenen Punkte `A`, `B` und `C` bis zu einer maximalen Rekursionstiefe `rek` zeichnet. Die Rekursionstiefe gibt an, wie oft die Funktion noch versucht in den Teildreiecken Sierpinski-Dreiecke zu zeichnen. Wird diese Funktion also mit einer Rekursionstiefe von `rek=0` aufgerufen, zeichnet sie einfach ein Dreieck durch die Punkte. Ansonsten zeichnet sie ein Sierpinski-Dreieck in den drei untergeordneten Dreiecken (oben, links, rechts), allerdings mit *um 1 reduzierter Rekursionstiefe*.

## Teil B: Zelluläre Automaten in einer Dimension

Nachdem wir zuerst eine graphische Darstellung des Sierpinksi-Dreiecks im vorherigen Aufgabenteil erarbeitet haben, möchten wir in diesem zweiten Teil auf die zeitlichen Entwicklung eines elementaren **1D zellulärer Automaten** eingehen. Ein solcher Automat ist ein diskretes System aus einzelnen Zellen, die alle in einer Linie angeordnet sind und die Zustände `0` und `1` annehmen können. Basierend auf dem aktuellen Zustand einer Zelle, sowie den Zuständen ihrer beiden direkt angrenzenden Nachbarzellen, wird in jedem Zeitschritt der Zustand jeder Zelle aktualisiert. 
Weil jede Zelle nur zwei Zustände annehmen kann und damit in Kombination mit ihren beiden Nachbarn nur 8 verschiedene Möglichkeiten denkbar sind, die eine Zelle beim Berechnen ihres neuen Zustandes zu Grunde liegen, charakterisierte Stephen Wolfram die Menge aller denkbarer elementarer, zellulärer Automaten gemäß der Binärdarstellung ihrer Update-Vorschriften.

1. Definieren Sie den aktuellen Zustand des Automaten als ein Array aus Integers mit der Länge `40`.
2. Als Anfangsbedingung können Sie alle Elemente des Arrays auf `0` setzen, und nur das mittlere Element auf `1`
3. Implementieren Sie das Update gemäß Regel 90 und führen Sie `50` Update-Schritte aus. Geben Sie nach jedem Schritt den Zustand Ihres Automaten aus
4. Vergleichen Sie ihre Ausgabe mit dem Sierpinski-Dreieck aus Teil A!
