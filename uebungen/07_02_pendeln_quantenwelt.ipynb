In dieser Aufgabe betrachten wir die Zeitentwicklung von Wellenpaketen im **quantenmechanischen harmonischen Oszillator**. Im Folgenden wollen wir diesen Oszillator zuerst analytisch lösen, um in einem zweiten Schritt jedes Wellenpaket in Basiszustände zu zerlegen und diese in der Zeit zu entwickeln.

Der harmoische Oszillator wird durch ein Teilchen in einem Potential 
    $V(x) = \omega^2 x^2/(2m)$ 
mit der Frequenz $\omega$ und der Teilchenmasse $m$ beschrieben. Wir setzen der Einfachheit halber $\hbar = \omega = m = 1$. Der harmonische Oszillator wird also im allgemeinen durch eine *zeitabhängige* Schrödinger-Gleichung der Form 

\begin{equation}
\frac{1}{2}\left[-\frac{\partial^2}{\partial x^2} + x^2\right]\psi(x,t) = i\frac{\partial}{\partial t}\psi(x,t) 
\end{equation}

beschrieben.

Mit dem Produktansatz $\psi(x,t)=\phi(x)\chi(t)$, d.h. der Separation von Orts- und Zeitkoordinate, reduziert sich die Schrödinger-Gleichung auf *zwei gewöhnliche* Differentialgleichungen. 
Die Lösungen für den Ortsanteil lassen sich über die sogenannten **Hermite-Polynome** $H_n$ darstellen:

\begin{equation}
\phi_n(x) =\pi^{-\frac{1}{4}} \frac{1}{\sqrt{2^n n!}} H_n(x)
e^{-\frac{x^2}{2}}, \qquad \text{mit } n=0,1,2,\ldots
\end{equation}

Die Hermite-Polynome $H_n(x)$ können mit folgender Vorschrift rekursiv konstruiert werden:

\begin{align}
H_0(x) &= 1,\\
H_1(x) &= 2x, \\
H_{n+1}(x) &= 2x H_n(x)-2nH_{n-1}(x).
\end{align}

Die Lösung für den Zeitanteil der Schrödinger-Gleichung hat die Form

\begin{equation}
\chi_n(t)=e^{-iE_nt},
\end{equation}

wobei die Energie des $n$-ten Zustands gegeben ist als $E_n = n + \tfrac{1}{2}$. 

Beachten Sie, dass die Funktionen $\psi_n(x,t) = \phi_n(x)\chi_n(t)$ lediglich eine Basis des Lösungsraums der Schrödingergleichung bilden, d.h. wir können eine allgemeine Lösung als Linearkombination dieser Basiselemente schreiben:

\begin{equation}
\psi(x,t) = \sum_{n=0}^{\infty}c_n \phi_n(x) \chi_n(t).
\end{equation}

Wir wollen nun für eine beliebige vorgegeben Wellenfunktion die Koeffizienten $c_n$ berechnen. Dafür machen wir uns die Tatsache zunutze, dass die Funktionen $\psi_n$ eine Orthogonalbasis darstellen. Das Skalarprodukt $\langle\cdot|\cdot\rangle$ auf dem Raum der Lösungen ist für zwei beliebige Wellenfunktionen $\psi_n$ und $\psi_m$ gegeben als

\begin{equation}
\langle\psi_n(x,t)|\psi_m(x,t)\rangle =
\int_{-\infty}^{\infty}\text{d}x\, \psi^*_n(x, t)\psi_m(x, t).
\end{equation}

Bekanntermaßen gilt für das Skalarprodukt zweier orthonormaler Basisvektoren immer

\begin{equation}
\langle\psi_n(x,t)|\psi_m(x,t)\rangle = \delta_{n,m},
\end{equation}

d.h. wir können die Koeffizienten $c_n$ des allgemeinen Zustands $\psi(x,t)$ durch Projektion des Zustands auf die Basisvektoren finden:

\begin{equation}
c_n = \langle\psi_n(x,t)|\psi(x,t)\rangle.
\end{equation}


#### Aufgaben

Ihre Aufgabe ist es numerisch die **Zeitentwicklung** der beiden Wellenfunktionen (gegeben bei $t=0$) zu berechnen:

\begin{align}
\psi_A(x,t=0) &= \pi^{-1/4}e^{-\frac{1}{2}(x-2)^2}, \\
\psi_B(x,t=0) &= \frac{1}{\pi}\frac{0.05}{(x-2)^2+0.05^2}.
\end{align}

Letzten Endes sollen Sie eine Animation für das Betragsquadrat $|\psi_{A/B}(x,t)|^2$ sowie den Real- und Imaginärteil erstelen. In der Aufgabe können Sie wie folgt vorgehen:
1. Implementieren Sie die Eigenfunktionen $\phi_n(x,t)$ in Julia als Funktionen
2. Implementieren Sie die Wellenpakete $\psi_{A/B}(x,t=0)$ in Julia als Funktionen
3. Implementieren Sie eine Funktion, die die Koeffizienten $c_n$ mittels Skalarprodukt (Integration) berechnet. Sie können sich hier auf die ersten zehn Basiszustände beschränken ($n=10$). Für die Integration können Sie entweder ein Paket Ihrer Wahl verwenden, oder Ihre Integration von bisherigen Aufgabenblättern verwenden. Wir schlagen vor, die Funktionen im Intervall $x\in[-6.0, 6.0]$ für etwa 1000 Stützstellen zu integrieren.
4. Nach Berechnung der Koeffizienten $c_n$ können Sie die Wellenfunktion $\psi(x,t) = \sum_{n=0}^{\infty}c_n \phi_n(x) \chi_n(t)$ berechnen und im Zeitintervall $t\in[0,10]$ als Animation darstellen.
