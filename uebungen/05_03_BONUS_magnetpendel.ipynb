Wir betrachten ein **Pendel**, das nun magnetisch sei und in einer Ebene schwingt. Es wechselwirke mit dem **Schwerefeld** der Erde, sowie mit **drei Magneten** und werde durch **Reibung** gebremst. Die drei Magnete seien gleichmäßig auf einem Ring mit Radius $d = 1$ um den Ursprung angeordnet, wie auf der Skizze gezeigt.



Die gravitative Wechselwirkung nähern wir wie zuvor durch

\begin{equation}\vec{F}_g \approx m\,g\,h\;\vec{e}_z \approx k_g \vec{r} \,.\end{equation}

Die Konstante $k_g$ steuert die Stärke der Gravitation und $\vec{r}$ ist der aktuelle Ortsvektor des Pendels. Die Näherung ist, wie auch beim eindimensionalen Pendel, dass wir nur kleine Auslenkungen aus der Ruhelage betrachten.

Kommen wir nun zu den drei Magneten, welche an Positionen $\vec{r}_i$ montiert sind. Sie üben eine Kraft $\vec{F}_{k,i}$ auf das Pendel aus, wobei $i$ der Index des Körpers $\{1,2,3\}$ ist und $k_i$ eine weitere Kraftkonstante. So ergibt sich das folgende Kraftgesetz:

\begin{equation}\vec{F}_{k, i} = k_i\dfrac{\vec{r}-\vec{r}_i}{|\vec{r}-\vec{r}_i|^3} \,.\end{equation}


Zuletzt fügen wir noch eine Reibungskraft $\vec{F}_r$ hinzu, die proportional zur Geschwindigkeit ist und deren Stärke durch $\gamma$ bestimmt wird

\begin{equation}\vec{F}_r = -\gamma\vec{v} \,.\end{equation}

Lassen Sie sich nicht durch die Anzahl der Konstanten abschrecken. Die Simulation ist nicht viel schwerer als die vorherigen Aufgaben! Die Gesamtbewegungsgleichung ist also

\begin{equation}
\vec{a}(t) = \vec{F}_g + \sum\limits_{i} \vec{F}_{k, i} + \vec{F}_r
= -k_g \vec{r} + \sum\limits_{i} k_i\dfrac{\vec{r}-\vec{r}_i}{|\vec{r}-\vec{r}_i|^3} - \gamma\vec{v} \,,
\end{equation}

wobei wir die Masse des Pendels $m = 1$ gesetzt haben und $a(t)$ die Beschleunigung bezeichnet. Wir nehmen an, dass die Auslenkung des Pendels in $z$-Richtung vernachlässigbar ist und betrachten dadurch effektiv ein zweidimensionales Problem, d.h. Sie berechnen nur die $x$ und $y$ Komponenten des Orts, der Geschwindigkeit und der Beschleunigung. Beachten Sie aber, dass die Ebene in der das Pendel schwingt im Abstand $h$ über der Ebene der Magneten liegt. In der Berechnung der Abstandsvektoren zwischen Pendel und Magneten geht der Abstand $h$ also sehr wohl ein.

Die passende Wahl des Algorithmus zur Integration der obigen Bewegungsgleichung sei Ihnen selbst überlassen. Der bereits behandelte Runge-Kutta-Algorithmus ist hierfür gut geeignet, aber Sie können auch einmal probieren, ob der backward-Euler Algorithmus stabil genug ist. Sie haben außerdem den Verlet-Algorithmus kennengelernt, der speziell für das Lösen von Bewegungsgleichungen entwickelt wurde.

1. Beginnen Sie damit, Trajektorien des Pendels zu berechnen, und stellen Sie diese grafisch dar. Hierzu müssen Sich sich insbesondere geeignete Abbruchbedingungen überlegen. Welche Endpositionen sind überhaupt möglich?
2. In einem zweiten Schritt zeichnen Sie nun eine Karte, die jeder Startposition eine Farbe zuordnet, welche einer der möglichen Endpositionen entsprechen soll. Für die Magneten als Endposition können Sie z.B. die Farben so wählen wie in der Skizze. Beachten Sie, dass dieser Schritt je nach Algorithmus und Abbruchbedingung aus 1. mehrere Minuten für ein Bild der Größe 100x100 Pixel betragen kann.

Als gute Werte für die oben genannten Konstanten haben sich herausgestellt:
$k_g = 0.2,\quad k_{1} = k_{2} = k_{3} = 20.0,\quad \gamma = 0.5,\quad h = 0.2$
