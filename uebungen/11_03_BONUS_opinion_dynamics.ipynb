In dieser Aufgabe wollen wir uns mit dem Thema **Meinungsdynamik** und Bildung von Konsens aus dem Bereich der Soziophysik und Politikwissenschaften befassen. Dabei betrachten wir die Interaktion verschiedener Individuen, die sich austauschen und dadurch gegenseitig ihre Meinung beeinflussen können. Durch die Details dieser Interaktion können verschiedene Meinungsverteilungen innerhalb der gesamten Gruppe simuliert werden.



### Multi-Agenten Simulationen für Meinungsdynamik

Im Gegensatz zu den bisher angewandten Algorithmen zum Lösen von Differentialgleichungen oder dem statistischen Integrieren, möchten wir in dieser Aufgabe das Konzept einer **Multi-Agenten Simulation** einführen und benutzen. Im Kern einer solchen Simulation steht eine Gruppe von sogenannten Agenten welche sich durch ihre **Eigenschaften** (meist interne Variablen) charakterisieren. In unserem Fall besteht die Eigenschaft jedes Agenten aus seiner Meinung im Bezug auf eine bestimmte Hypothese, die wir im Folgenden als reelle Zahl $a_i \in [-M,M]$ beschreiben möchten, wobei $M$ das Maximum der Meinung darstellt. Die Gesamtmenge der Agenten beschreibt eine soziale Gruppe, und ähnlich wie wir von kleinen physikalischen Systemen auf den thermodynamischen Limes extrapolieren, ist auch hier die Annahmen, dass die Gruppe von Agenten im weitesten Sinne eine gut durchmischte Gesellschaft beschreibt. 

Neben den Eigenschaften jedes Agenten gehört zu einer Multi-Agenten Simulation noch eine Menge an **Verhaltensregeln** nach denen sich die Agenten zu diskreten Zeitschritten verhalten. In unserem Fall wollen wir die folgenden Verhaltensregeln annehmen:
- In einem Zeitschritt wird jeder Agent $i$ mit der Meinung eines zufällige ausgewählten, anderen Agenten $j$ konfrontiert. (Dies modelliert z.B. das Agent $j$ dem Agenten $i$ von seiner Meinung erzählt oder Agent $i$ die Meinung von Agent $j$ irgendwo liest)
- Durch die Meinungskonfrontation ändert sich die Meinung des Agenten $i$ von $a_i$ zu $a_i + \alpha f(a_i, a_j)$ wobei die Konstante $\alpha$ die Stärke des Austauschs parametrisiert und die Funktion $f$ beschreibt, wie sehr die Meinung $a_j$ den Agenten mit Meinung $a_i$ beeinflusst. *Hinweis*: Die Meinung jedes Agenten ist auf das Intervall $a_i \in [-M,M]$ festgelegt, d.h. Sie kann sich nicht aus diesem Intervall hinaus verändern.
- Unabhängig vom Austausch zwischen Agenten kann sich ein Agent mit einer Wahrscheinlichkeit $\theta$ eine gänzlich neue Meinung erhalten (Dies modelliert z.B. das der Agent sich selber etwas überlegt, ohne von anderen beeinflusst zu werden). 

Aus diesen Verhaltensregeln wird deutlich, dass neben der Wahl der Parameter $\alpha$ (Beeinflussungsstärke) und $\theta$ (Selbstbezogenheit) vor allem die genaue Form der Funktion $f$ entscheidend für die Simulation ist. Im Folgenden wollen wir verschiedene Szenarien mit unterschiedlichen Funktionen $f$ diskutieren und so auf verschiedene gesellschaftliche Phänomene eingehen.

Im Folgenden sollen Sie zunächst die Grundkonstruktion einer Multi-Agenten Simulation in Julia vornehmen. Dafür benötigen Sie die folgenden einzelnen Bausteine / Funktionen:

3. Plotten einer gegebenen Zeitentwicklung. Dies können Sie entweder machen, indem Sie die Meinung jedes Agenten mittels `plot` darstellen und mittels `alpha=0.05` die Sichtbarkeit der einzelnen Linien reduzieren, oder Sie plotten ein zu jedem Zeitpunkt selbst-erstelltes Histogram der Meinung über die Zeit mittels `imshow`.

Implementieren Sie im Folgenden diese drei Bausteine als Julia Funktionen um im Folgenden verschiedene Beispiele betrachten zu können.

### 1. Extreme Beeinflussung, "Contagion" und die Bildung von extremem Konsens

Das erste Beispiel, das wir uns anschauen wollen ist das Beispiel von extremer Beeinflussung. Bei diesem Prozess, der im Englischen den Namen **Contagion** trägt, ändert ein Agent seine Meinung in die Richtung der Meinung des beeinflussenden Agenten. In Formeln bedeutet dies

\begin{equation}
f(a_i, a_j) = a_j
\end{equation}

Dies bedeutet, dass ein Agent immer in die Richtung der Meinung beeinflusst wird, die er hört. Weiterhin wird er selbst noch extremer in seiner Meinung, selbst wenn er eine weniger starke Meinung der eigenen Richtung vernimmt.

Simulieren Sie Ihr Multi-Agenten Modell für $100$ Individuen mit der angegebenen Funktion $f$, $\alpha=0.05$ sowie $M=3.5$ für $N_t = 200$ Zeitschritte. Vernachlässigen Sie zuerst die Selbstbestimmung der Agenten (d.h. setzen Sie $\theta=0$) und beschreiben Sie welches Phänomen Sie in ihrer Simulation erkennen können. Zeigen Sie, dass dieses Phänomen selbst für geringe Maße von Selbstbestimmung (z.B. $\theta=0.01$) robust ist. 

### 2. Kompromisse, "Assimilation" und neutraler Konsens

Im nächsten Beispiel schauen wir uns an, wie sich das Verhalten der Gruppe verändert, wenn jeder Agent nur noch Kompromisse als Änderung seiner Meinung zulässt. Die Änderungsfunktion $f$ wird somit zu

\begin{equation}
f(a_i, a_j) = a_j - a_i
\end{equation}

Dies bedeutet, dass ein Agent immer in die Richtung der Meinung beeinflusst wird, die den Kompromiss zwischen seiner und der beeinflussenden Meinung darstellt.

Wiederholen Sie Ihre Simulation für $100$ Individuen mit der angegebenen Funktion $f$, $\alpha=0.05$ sowie $M=3.5$ für $N_t = 200$ Zeitschritte und eine Selbstbestimmung $\theta=0.01$. Was können Sie beobachten? 

### 3. Kompromisse vs. extreme Beeinflussung und die Rolle der Selbstbestimmung

Nachdem wir uns zunächst Kompromissverhalten, sowie beeinflusstes Verhalten gesondert angeschaut haben, möchten wir nun beide Verhalten in einer Funktion mixen. Dafüer definieren wir $f$ für einen Kompromissgrad $\rho \in [0,1]$ als

\begin{equation}
f(a_i, a_j) = a_j - \rho a_i
\end{equation}

Wir wollen jetzt untersuchen, welcher Effekt stärker ist: Die Kompromissbereitschatft ($\rho=1$) oder die extreme Beeinflussung ($\rho=0$). Dafür wählen wir zunächst $\theta=0$ und betrachten die Fälle $\rho=0.97$ und $\rho=1$ mit den Parametern $\alpha=0.5$, $N=100$, $M=3.5$ sowie $N_t=500$.

Beschreiben Sie die Rolle von $\rho$ im Bezug auf die Meinungsbildung innerhalb der Gruppe.

Untersuchen Sie jetzt was für eine Rolle die Selbstbestimmung $\theta$ an dieser Stelle spielt. Wählen Sie dafür ein $\rho=0.9$ (welches nach Ihren Ergebnissen vorher eine deutliche Tendenz zu einem Gruppenverhalten ausbilden sollte) und erhöhen Sie $\theta$ von $0$ auf $0.1$.

Was können Sie im Meinungsverhalten der Gruppe bei steigender Selbstbestimmung beobachten? 

### 4. Meinungsbildung unter beschränktem Vertrauen und die Bildung von fragmentierten Gesellschaften

Bisher sind alle Modelle, die wir untersucht haben, davon ausgegangen, dass Agenten einander unabhängig ihrer Meinungsdistanz beeinflussen. Empirische Studien in unserer Gesellschaft zeigen allerdings das Gegenteil, es scheint ein größerer Trend dahin zu gehen, dass die Beeinflussbarkeit der eigenen Meinung abnimmt, je weiter entfernt die beeinflussende Meinung von der eigenen abweicht. Diesen Effekt nennt man **beschränktes Vertrauen** (engl. *motivated cognition*). Im Folgenden wollen wir seinen Einfluss auf die Meinungsdynamik exemplarisch untersuchen. 

Eine einfache Möglichkeit, beschränktes Vertrauen in die Multi-Agenten Simulation einzubauen, ist die Funktion $f$ um einen Vertrauensfaktor zu erweitern. Dieser Vertrauensfaktor ist mittels zweier Parameter $\lambda$ und $k$ definiert als

\begin{equation}
\frac{\lambda^k}{\lambda^k + |a_j - a_i|^k}
\end{equation}

Dabei beschreibt $\lambda$ die Breite des Vertrauensintervalls, d.h. wie weit entfernt Meinungen in die Bildung der eigenen Meinung mit einbezogen werden, und $k$ ist die Stärke des Abfalls des Vertrauens. Dabei bedeuten höhere Werte $k$ ein stärkeres Abfallen des Vertrauens außérhalb des Intervalls $\lambda$.

Die gesamte Funktion $f$ lässt sich somit schreiben als

\begin{equation}
f(a_i, a_j) = \frac{\lambda^k}{\lambda^k + |a_j - a_i|^k}  (a_j - \rho a_i)
\end{equation}

Plotten Sie $f$ um sich mit den verschiedenen Parametern vertraut zu machen und untersuchen Sie danach die Rolle von beschränktem Vertrauen auf die Bildung von Meinungen innerhalb der Gesellschaft. Ein guter Ansatz dafür ist es mit $k=2$ und $\lambda \gg M$ eine vorherige Simulation zu wiederholen und graduell $\lambda$ zu verringern.

Desweiteren können Sie auch folgende Parametersätze verwenden:
- Vertrauensintervall $\lambda = 0.5$, $k=2$, Selbstbestimmung $\theta=0.01$, Kompromissstärke $\rho=0$
- Vertrauensintervall $\lambda = 0.5$, $k=2$, Selbstbestimmung $\theta=0.01$, Kompromissstärke $\rho=1$
- Vertrauensintervall $\lambda = 0.5$, $k=10$, Selbstbestimmung $\theta=0.01$, Kompromissstärke $\rho=1$
- Vertrauensintervall $\lambda = 1.7$, $k=10$, Selbstbestimmung $\theta=0.01$, Kompromissstärke $\rho=1$

Beschreiben Sie jeweils Ihre Beobachtungen.
