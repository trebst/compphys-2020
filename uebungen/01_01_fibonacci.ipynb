Für die Implementierung in dieser Aufgabe möchten wir das Konzept eines **Arrays** oder einer **Liste** verwenden. Grundsätzlich versteht man unter einem Array eine Container-Variable mit Komponenten gleichen Typs, die alle mit einem Index referenziert werden können. Es werden also im Prinzip viele einzelne Variablen zu einem Array zusammengefasst. 

So wie Sie z.B. in der Mathematik ein Objekt $x$ mit Komponenenten $x_1, \;\dots, x_n$ definieren können, ist es Ihnen auch in Julia möglich eine Variable `x` mit Komponenten Variable `x[i]` zu definieren. 



Syntaktisch können Sie ein Array aus `n` Komponenten entweder mittels

    x = zeros(n)
    
erstellen (dies erstellt `n` Nullen des Typs `Float64`), oder mittels

    x = Vector{Int64}(undef, n)

wobei Sie damit `n` nicht definierte (undef) Komponenten eines festen Typs (hier `Int64`) erzeugen, die Sie später noch einzeln setzen müssen.

### b) Fibonacci-Zahlen

Die **Fibonacci Zahlen** sind definiert als die Folge der Zahlen $\{F_n \; \forall n \in \mathbb{N}\}$, die folgende Relation erfüllt:

$\quad F_n = F_{n - 1} + F_{n - 2}$
mit den Anfangswerten $F_1 = 1$ und $F_2 = 1$.

Diese mathematische Definition ergibt eine praktische Vorschrift die Zahlen nacheinander zu berechnen. Implementieren Sie im Folgenden einen kleinen Julia Code, der die ersten 15 Fibonacci-Zahlen berechnet, in einer Liste speichert und ausgibt.
